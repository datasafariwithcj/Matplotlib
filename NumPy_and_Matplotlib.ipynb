# Grid, Axes, and Labels

```
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, -2**x)
plt.grid(True)
plt.axis([-5,5,-5,5])
plt.show()
```

```
x = np.arange(3)
plt.plot(x, -x**2, x, -x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2','-x**3','-2*x','-2**x'], loc='upper left')
plt.grid(True)
plt.xlim([-5,+5])
plt.ylim([-5,+5])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Data Safari with CJ')
print(plt.axis())
plt.show()
```

# Colours, Styles and Markers

```
x = np.arange(5)
y = x
plt.plot(x, y+0.2, 'g')
plt.plot(x, y, 'r')
plt.plot(x, y-0.2, 'c')
plt.plot(x, y-0.4, 'k')
plt.plot(x, y-0.6, 'm')
plt.plot(x, y-0.8, 'w')
plt.plot(x, y-1, 'b')
plt.show()
```

# Subplots

```
x = np.arange(3)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

plt.subplot(2,2,1)
plt.plot(x, -x**2)

plt.subplot(2,2,2)
plt.plot(x, -x**3)

plt.subplot(2,2,3)
plt.plot(x, -2*x)

plt.subplot(2,2,4)
plt.plot(x, -2**x)

plt.show()
```

```
fig, axs = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.3,hspace=0.3)

axs[0, 0].plot(x, -x**2)
axs[0, 1].plot(x, -x**3)
axs[1, 0].plot(x, -2*x)
axs[1, 1].plot(x, -2**x)

plt.show()
```

```
n = 3
data_for_plot = np.linspace(0, 2 * np.pi * n, 300)

def trig_func(sty_name):
    plt.style.use(sty_name)
    plt.plot(data_for_plot, np.sin(data_for_plot), label = 'sin')
    plt.plot(data_for_plot, np.cos(data_for_plot), label = 'cos')
    plt.legend()
    plt.show()

trig_func('default')
```

```
from posixpath import split
fig = plt.figure(constrained_layout = True)
splits = gridspec.GridSpec(3, 3, fig)

ax1 =  fig.add_subplot(splits[0, :])
ax1.set_title('[0, :]')
ax2 =  fig.add_subplot(splits[1, :2])
ax2.set_title('[1, :2]')
ax3 =  fig.add_subplot(splits[1:, 2])
ax3.set_title('[1, 2]')
ax4 = fig.add_subplot(splits[2, 0])
ax4.set_title('[2, 0]')
ax5 = fig.add_subplot(splits[2, 1])
ax5.set_title('[2, 1:]')

plt.show()
```
